# 1. Load xls sheet using pandas

```python
# Load xls sheet with data
file_path = '../Data/nanonose.xls'
doc = pd.read_excel(file_path)
```

```python
# Print data to get ann overview
doc.head(10)
```

# 2. Assign useful and important variables

## Attribute names

```python
# Extract attribute names (1st row, column 4 to 12)
attributeNames  = doc.columns[3:11].tolist()
attributeNames
```

## Define y

```python
# Extract class names to python list
# by removing the 2nd (empty) row and renaming it to 'sample type'
# classColumnName = doc['Nanonose'].iloc[0]
y = doc[['Nanonose']].iloc[1:].reset_index(drop=True)
y.rename(columns = {'Nanonose': 'class'}, inplace = True)
y
```

## Class names

```python
classNamesList = sorted(y['class'].unique())
classNamesList
```

```python
classDict       = dict(zip(classNamesList, range(len(classNamesList))))
classDict
```

```python
classNames = y[['class']].drop_duplicates().reset_index(drop=True)
classNames
```

## Define X

```python
# Define X
# by dropping the entire first row (and resetting index)
# and then only chosing column names that have attribute names
X = doc[attributeNames].iloc[1:].reset_index(drop=True)
X
```

```python
# Summary statistics about X data'
X.describe()
```

```python
if y.shape[0] == X.shape[0]:
    print('Good job!\n\ty and X have same dimensions!')
else:
    print('Double check y and X!')
    print('\ty shape: {}\nX shape: {}'.format(y.shape, X.shape))
```

## Dimensionality constants
* Number of observations *N*
* Number of Attributes *M*
* Number of classes *C*

```python
# Compute values of N, M and C.
N = len(y)              # number of observations
M = len(attributeNames) # number of attributes
C = len(classNames)     # number of classes
```
